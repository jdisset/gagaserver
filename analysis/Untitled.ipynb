```python
DB_FILENAME = "../build/save.db"

import json
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.context('Solarize_Light2')
```

```python
conn = sqlite3.connect(DB_FILENAME)
runs = pd.read_sql_query("SELECT * FROM run", conn)
runs

def get_scalar_result(sql, conn):
    cursor=conn.cursor()
    cursor.execute(sql)
    return cursor.fetchone()[0]
```

```python
q = """
SELECT individual.id, value, number as gen
FROM individual, evaluation, generation
WHERE evaluation.id_individual = individual.id 
AND individual.id_generation = generation.id 
AND id_generation IN (SELECT id FROM generation WHERE id_run = 1 AND number < 5)
"""
q = """
SELECT individual.id, number as gen
FROM individual, generation
WHERE individual.id_generation = generation.id 
AND id_generation IN (SELECT id FROM generation WHERE id_run = 1 AND number < 5)
"""
individuals = pd.read_sql_query(q, conn)
indList = individuals['id'].unique().tolist()
idToGen = {row['id'] : row['gen'] for index,row in individuals.iterrows()}
idToGen
```

```python
import networkx as nx
G = nx.DiGraph()
G.add_nodes_from(indList)
```

```python
inheritance = pd.read_sql_query("SELECT * FROM inheritance WHERE id_child IN (SELECT individual.id FROM individual, generation WHERE individual.id_generation = generation.id AND generation.number < 5)", conn)
for index, row in inheritance.iterrows():
    c = 'black' if row['type'] == 'C' else ('blue' if row['type'] == 'X' else 'orange')
    G.add_edge(row['id_parent'], row['id_child'], color=c)
```

```python
nx.write_gml(G, "graph.gml")
```
